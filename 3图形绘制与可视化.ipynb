# Python-Matplotlib 绘图基础

## 基本概念

### figure 画布

画图板/画布：这是一个基础载体，类似实际的画图板，用pyplot.figure()函数创建，程序中允许创建多个画图板，具体操作的画板遵循就近原则（操作是在最近一次调用的画图板上实现），缺省条件下内部默认调用pyplot.figure(1)。

### axes 绘图区

用来绘图的实际区域，一般不直接获取，直接设定方式为pyplot.axes([x, y, w, h])，即axes函数直接确定了该区域在画图板/画布中的位置为x,y 尺寸为w,h axes() VS subplot() pyplot.axes([x, y, w, h])是用来在画图板上确认图形区的位置和大小的函数，x,y表示图形区左下角相对于画图板的坐标（距离画布最左边x，画布最下边y），w,h表示图形区的宽高。（缺省时该操作在figure(1)上操作）

axes() VS axis()

axes([x, y, w, h])用来设定图形区

axis([x_left, x_right, y_bottom, y_top])是用来设置所绘制图形的视窗大小的，表示直接展示的图形是需要满足参数中范围的值，直观表现是绘图区实际展示的坐标范围。

### subplot 绘图区（分区）

pyplot.subplot(abc)本质也是用来确认图形区在画图板上位置大小的函数，区别是该函数将画图板按a行b列等分，然后逐行编号，并选择编号为c的区域作为图形区用来绘图。这是一个axes()操作的高级封装，方便用户使用。subplot(233)表示2行3列的第3个位置（即，第1行第三个区域） 注：subplot重复操作时如果前后两次的划分方式不一致时，会做必要的历史图形清理。


pyplot.subplot(abc) 当abc有小数时要加“,”

 subplot() VS plot() 

subplot用来生成图形区 plot是实际使用的绘图函数，类似的函数还有hist等，plot操作遵守就近原则，即作用在最近一次使用的图形区上。

### show展示

pyplot.show()实际展示的区域是画图板上所有图形区的最小包围区，不是整个画图板，即如果仅仅调用了subplot(224)结果只展示右下角的4号区域，而不是1、2、3、4都展示，因此会存在一定的错觉。

### 标签区

标签区：用来展示图形相关标签的地方，该区域根据图形区进行扩展，与该区域有关联的函数是pyplot.xlabel()、pyplot.ylabel()、pyplot.title()等

#### 点、线样式

点属性：

点型（标记marker）： 像素： ',' 圆形： 'o' 上三角： '^' 下三角： 'v' 左三角： '<' 右三角： '>' 方形： 's' 加号： '+' 叉形： 'x' 棱形： 'D' 细棱形： 'd' 三脚架朝下： '1'（就是丫） 三脚架朝上： '2' 三脚架朝左： '3' 三脚架朝右： '4' 六角形： 'h' 旋转六角形： 'H' 五角形： 'p' 垂直线： '|' 水平线： '_'

标记大小（markersize 简写为 ms）： markersize： 实数

标记边缘宽度（markeredgewidth 简写为 mew）： markeredgewidth：实数

标记边缘颜色（markeredgecolor 简写为 mec）： markeredgecolor：颜色选项中的任意值

标记表面颜色（markerfacecolor 简写为 mfc）： markerfacecolor：颜色选项中的任意值

透明度（alpha）： alpha： [0,1]之间的浮点数

线宽（linewidth）： linewidth： 实数

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
y = np.arange(1,5)  #默认x 1-3递增
plt.plot(y,marker='o')  #指定marker时，会画出线段
plt.plot(y+1,marker='D',linewidth=3)  #线宽
plt.plot(y+2,'^',markersize=10)  #标记大小
plt.plot(y+3,'p',ms=5)   #标记大小简写
plt.show()
```

线型（linestyle 简写为 ls）：

实线： '-' ///虚线： '--' ///虚点线： '-.' ///点线： ':'

```python
#四种线形 - 实线； -- 虚线；、-. 点划线 : 点线；
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
y = np.arange(1,5)
plt.plot(y,ls='-')  #实线
plt.plot(y+1,'--')
plt.plot(y+2,'-.')
plt.plot(y+3,'.')
plt.show()
```

### axes绘图区(嵌套)

add_axes([x,y,w,h])新增子区域,该区域可以座落在figure内任意位置，且该区域可任意设置大小

```python
import numpy as np  
import matplotlib.pyplot as plt  

#新建figure
fig = plt.figure
#定义数据
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 3, 4, 2, 5, 8, 6]
#新建区域ax1
  #figure的百分比，从figure 10%的位置开始绘制，宽高为80%
left,ottom,width,height = 0.1,0.1,0.8,0.8
  #获得绘制的句柄
ax1 = plt.axes([left,ottom,width,height])
ax1.plot(x,y,'r')
ax1.set_title('area1')  # 图表标题

#新建区域ax2,嵌套到ax1里
left,ottom,width,height = 0.2,0.6,0.25,0.25
  #获得绘制的句柄
ax1 = plt.axes([left,ottom,width,height])
ax1.plot(x,y,'b')
ax1.set_title('area2')

plt.show()
```

## axis坐标轴范围

### 坐标轴范围

axis([x_left, x_right, y_bottom, y_top])是用来设置所绘制图形的视窗大小的，表示直接展示的图形是需要满足参数中范围的值，直观表现是绘图区实际展示的坐标范围。注意，pyplot及划分的子图均有该函数。 xlim(xmin, xmax)和ylim(ymin, ymax)来调整x,y坐标范围。注意，只有pyplot有此函数，子图不可以这样设置

axis('square') 作图为正方形，并且x,y轴范围相同

axis('equal') x,y轴刻度等长

axis('off') 关闭坐标轴 官网上贴出的其他的一些选项

```python
import numpy as np
import matplotlib.pyplot as plt
#新建figure
fig = plt.figure()
# 定义数据
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 3, 4, 2, 5, 8, 6]
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
# 获得绘制的句柄
ax1 = plt.axes([left, bottom, width, height])
ax1.plot(x, y, 'r')
ax1.set_title('area1')

ax1.axis([0,10,1,8])  #x，y轴范围
#ax1.axis('equal')  #x,y轴刻度等长
#plt.axis(0,9)     #设置x轴
plt.axis('on')  #off 是否开启坐标轴
plt.show()
```

## 图形标签

### xlabel、ylabel、title、legend

xlabel与ylabel（设置坐标轴标签）、title（标题）、legend（图例）

title()可同时在子图中显示中间、左侧、右侧3个标题。 函数签名为matplotlib.pyplot.title(label, fontdict=None, loc=None, pad=None, *, y=None, **kwargs) 参数作用及取值如下：

#label：类型为字符串，即标题文本。

#fontdict：类型为字典，控制文本的字体属性。默认值为： {'fontsize': rcParams['axes.titlesize'], 'fontweight': rcParams['axes.titleweight'], 'color': rcParams['axes.titlecolor'], 'verticalalignment': 'baseline', 'horizontalalignment': loc}   （依次为：字体大小、字体粗细、字体颜色、垂直对齐，水平对齐）

#loc：取值范围为{'left', 'center', 'right'}，默认值为rcParams["axes.titlelocation"]（'center'），即标题的位置。

#pad：类型为浮点数，默认值为default: rcParams["axes.titlepad"] (6.0)。即标题与子图的填充距离（内边距）。
 
#y：类型为浮点数，默认值为rcParams["axes.titley"] (None)。即标题在子图中的垂直距离，单位为子图高度的百分比，1.0在子图最顶部，默认值None则自动确定标题位置，避免与其他元素重叠。
///////////////////////////////////////////////////////////

关于位置 标题：'left', 'center', 'right' x轴：'left', 'center', 'right' y轴：'bottom', 'center', 'top' 

图例：lower/upper 与left/center/right 组合使用(这些位置也可以用数值0-10代替),具体包括：

best/ upper/ right/ upper left/ lower left/ lower right/ right center/ left center/ right center/lower center/ upper center /

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
x = [1,2,3,4,5,6,7,8]
y = [1,5,3,4,6,7,2,8]
plt.rcParams['font.sans-serif'] = 'simhei' # 设置显示为中文的黑体
plt.title('标签示例',fontsize=20)
plt.xlabel('X',fontsize=14)
plt.ylabel('Y',fontsize=14,loc='top')
plt.plot(x,y,color='m',label='班级排名')
plt.legend(loc='upper center',fontsize=14)  #loc='lower right'
plt.show()
```

### text给图形增加数据

pyplot.text(x, y, s, fontdict=None, withdash=False, **kwargs)

常用参数说明：

x，y：显示内容的坐标位置

s：显示内容

fontdict：一个定义s格式的dict

fontsize：字体大小

color：str or tuple, 设置字体颜色 ,单个字符候选项{‘b’, ‘g’, ‘r’, ‘c’, ‘m’, ‘y’, ‘k’, ‘w’}，也可以’black’,'red’等，tuple时用[0,1]之间的浮点型数据，RGB或者RGBA, 如: (0.1, 0.2, 0.5)、(0.1, 0.2, 0.5, 0.3)等

backgroundcolor：字体背景颜色

horizontalalignment(ha)：设置水平对齐方式，可选参数：left,right,center

verticalalignment(va)：设置垂直对齐方式 ，可选参数 ： ‘center’ , ‘top’ , ‘bottom’ ,‘baseline’

rotation(旋转角度)：可选参数为:vertical,horizontal 也可以为数字

alpha：透明度，参数值0至1之间



```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
x = [1,2,3,4,5,6,7,8]
y = [1,5,3,4,6,7,2,8]
plt.rcParams['font.sans-serif'] = 'simhei' # 设置显示为中文的黑体
plt.title('标签示例',fontsize=20)
plt.xlabel('X',fontsize=14)
plt.ylabel('Y',fontsize=14,loc='top')
for a,b in zip(x,y):     # b+0.2 text显示在点上面一点
    plt.text(a,b+0.2,b,fontsize=12,color=(0.1,0.2,0.9),backgroundcolor='y',rotation=30,alpha=0.5)
plt.plot(x,y,color='m',label='班级排名')
plt.legend(loc='lower center',fontsize=14)
plt.show()
```

## hist 直方图绘制

由一系列的高度不等的纵向或横向条纹或线段来表示数据的分布情况，适用于观察某一个变量在一个连续的区间的分布情况。 注意区分条形图，条形图也可以用来观察数据的分布情况，但使用于观察某个变量在不连续的区间的分布情况。

条形图：条形图用长条形表示每一个类别，长条形的长度表示类别的频数，宽度表示表示类别。 

直方图：直方图是一种统计报告图，形式上也是一个个的长条形，但是直方图用长条形的面积表示频数，所以长条形的高度表示频数组距频数组距，宽度表示组距，其长度和宽度均有意义。当宽度相同时，一般就用长条形长度表示频数。

参数参考：https://www.cnblogs.com/LiErRui/articles/11588399.html

pie()函数的参数如下： x：各个饼块的尺寸。类1维数组结构。 explode：每个饼块相对于饼圆半径的偏移距离，取值为小数。类1维数组结构。默认值为None。 labels：每个饼块的标签。字符串列表。默认值为None。 colors：每个冰块的颜色。类数组结构。颜色会循环使用。默认值为None，使用当前色彩循环。 autopct：饼块内标签。None或字符串或可调用对象。默认值为None。如果值为格式字符串，标签将被格式化，如果值为函数，将被直接调用。 pctdistance：饼块内标签与圆心的距离。浮点数。默认值为0.6，autopct不为None该参数生效。 shadow：饼图下是否有阴影。布尔值。默认值为False。 labeldistance：饼块外标签与圆心的距离。浮点值或None。默认值为1.1。如果设置为None，标签不会显示，但是图例可以使用标签。 startangle：饼块起始角度。浮点数。默认值为0，即从x轴开始。角度逆时针旋转。 radius：饼图半径。浮点数。默认值为1. counterclock：角度是否逆时针旋转。布尔值。默认值为True。 wedgeprops：饼块属性。字典。默认值为None。具体见matplotlib.patches.Wedge 。 textprops：文本属性。字典。默认值为None。 center：饼图中心坐标。(float,float)浮点数二元组。默认值为(0,0)。 frame：是否绘制子图边框。布尔值。默认为False。 rotatelabels：饼块外标签是否按饼块角度旋转。布尔值。默认为False。 normalize：是否归一化。布尔值或None。默认值为True。 True：完全饼图，对x进行归一化，sum(x)==1。 False：如果sum(x)<=1，绘制不完全饼图。如果sum(x)>1，抛出ValueError异常。 None：如果sum(x)>=1，默认值为True。如果sum(x)<1，默认值为False。 绘制不完全饼图，需要明确传递normalize=False。
