# 相关性分析

Pearson相关系数法求解相关系数矩阵 DataFrame.corr(method='pearson', min_periods=1)

参数说明：

method：可选值为{‘pearson’, ‘kendall’, ‘spearman’}

pearson：pearson correlation coefficient（皮尔逊相关性系数）。用来衡量两个数据集合是否在一条线上面，即针对线性数据的相关系数计算，针对非线性数据便会有误差。

kendall：kendall correlation coefficient（肯德尔相关性系数）。用于反映分类变量相关性的指标，即针对无序序列的相关系数，非正太分布的数据

spearman：spearman correlation coefficient（斯皮尔曼秩相关性系数）。非线性的，非正太分布的数据的相关系数。

所以针对【连续、正态分布、线性】数据，采用pearson相关系数；针对【非线性的、非正态】数据，采用spearman相关系数；针对【分类变量、无序】数据，采用Kendall相关系数。

一般来讲，线性数据采用pearson，否则选择spearman，如果是分类的则用kendall。

min_periods：int，可选参数，为了获得有效结果所需要的最小的观察数量

# 数据预处理

已经提取到了12个相关属性，使用lasso回归方法进行关键属性选取，降低模型的复杂程度

Lasso回归可以有效地对多重共线性特征进行筛选。 缺点是如果存在一组高度相关的特征时，Lasso回归方法倾向与选择其中一个特征，而忽视其他所有特征，这种情况可能会导致结果的不稳定。但在合适的场景中还是可以发挥不错的效果。

案例这里由于各原始属性存在着严重的多重共线性，这一问题已成为主要问题，这里采用Lasso回归进行关键属性选取是恰当的。 class sklearn.linear_model.Lasso(alpha=1.0, *, fit_intercept=True, normalize='deprecated', precompute=False, copy_X=True, max_iter=1000, tol=0.0001, warm_start=False, positive=False, random_state=None, selection='cyclic')

alpha为调整参数（非负正则参数），控制着模型的复杂程度，值越大，表示对特征较多的线性模型的惩罚力度就越大，从而最终获得一个特征较少的模型。 alpha的确定可以采用交叉验证法，选取交叉验证误差最小的值。在sklearn.linear_model中，可以使用LassoCV实现。

```python
data
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso(1000,max_iter=10000)   #第一个是alpha  max_iter迭代次数 默认1000
lasso.fit(data.iloc[:,0:13],data['y'])  #传入的数据 0:13左闭右开
print('相关系数：',np.round(lasso.coef_,5))  # lasso.coef_相关系数系列
```

```python
from sklearn.linear_model import Lasso,LassoCV,LassoLarsCV   
# Lasso回归,LassoCV交叉验证实现alpha的选取，LassoLarsCV基于最小角回归交叉验证实现alpha的选取
lasso=LassoCV()
lasso.fit(data.iloc[:,0:13],data['y'])
print('最佳的alpha：',lasso.alpha_)
```

## SVR算法

sklearn库的LinearSVR函数实现了线性支持向量回归，其使用语法如下： LinearSVR(C=1.0, dual=True, epsilon=0.0, fit_intercept=True,intercept_scaling=1.0, loss='epsilon_insensitive', max_iter=10000, random_state=None, tol=0.0001, verbose=0) max_iter 接收int。指定最大迭代次数，默认为1000。 由于支持向量机拥有完善的理论基础和良好的特性，人们对其进行了广泛的研究和应用，涉及分类、回归、聚类、时间序列分析、异常点检测等诸多方面。

其中 C  为惩罚系数，起到了约束 ϵ \epsilonϵ 的作用，C越大ϵ越小



优点：支持向量回归不仅适用于线性模型，对于数据和特征之间的非线性关系也能很好抓住；支持向量回归不需要担心多重共线性问题，可以避免局部极小化问题，提高泛化性能，解决高维问题；支持向量回归虽然不会在过程中直接排除异常点，但会使得由异常点引起的偏差更小。 缺点：计算复杂度高，在面临数据量大的时候，计算耗时长。

所以线性、非线性都适用，但不适用于大量数据运算

https://blog.csdn.net/qq_42554780/article/details/120569414

由于预测结果各值数量相差较大，将每个值进行数据标准化

在本次实验我们采用Z-score方法，其他方法见老师教案

Z-score标准化法

当我们遇到某个指标的最大值和最小值未知的情况时，或者有超出取值范围的离群数值的时候，就不再适宜计算极差了，此时我们可以采用另一种数据标准化最常用的方法，即Z-score标准化，也叫标准差标准化法。

经过Z-score标准化后，数据将符合标准正态分布，即将有约一半观察值的数值小于0，另一半观察值的数值大于0

得到y的思路：  首先将所有数据标准化后通过支持向量模型求出y的z分数，再利用z分数的公式倒推出y的预测值

```python
import matplotlib.pyplot as plt  #为画图做准备
from sklearn.svm import LinearSVR  #从支持向量模型中导入支持向量回归
data = new_reg_data
feature = ['x1', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x13']  # 由lasso模型得出的关键属性所在列
data_train=data.loc[range(2002,2022)].copy()  # 取出所有2022年前的原始数据建模，因为训练集必须是原始数据   copy()可以保护源代码不受伤
#数据准备
data_mean = data_train.mean()
data_std = data_train.std()
data_train = (data_train - data_mean)/data_std  # 将训练集中的所有数据进行标准化
x_train = data_train[feature].values #取出标准化后的x1-x13的所有数据作为 属性数据  .values序列结构
y_train = data_train['y'].values  #取出标准化后y的所有数据作为 标签数据     .values序列结构

#使用SVR模型预测
linearsvr = LinearSVR(max_iter=10000)  # 调用LinearSVR()函数  max_iter最大迭代次数 
linearsvr.fit(x_train,y_train)         #传入原始值标准化后的数据
x = ((data[feature] - data_mean[feature])/data_std[feature]).values  # 预测返回z-score
data['y_pred'] = linearsvr.predict(x) * data_std['y'] + data_mean['y']  #并还原结果  --增加一列y_pred到data中
outputfile = '../chapter4Fiscalrevenueprojections/data/new_reg_data_GM11_SVR_1.xlsx'  # SVR预测后保存结果的地址
data.to_excel(outputfile)

print('真实值与预测值分别为：\n',data[['y','y_pred']])

fig = data[['y','y_pred']].plot(style=['bo-','r*-'])  # 画出预测结果图
#fig = data[['y','y_pred']].plot(subplots=True,style=['bo-','r*-'])  # 画出预测结果图
plt.show()
```

1 平均绝对误差 MAE 是所有单个观测值与算术平均值的偏差的绝对值的平均。 平均绝对误差可以避免误差相互抵消的问题，因而可以准确反映实际预测误差的大小。

2 均方根误差RMSE（Root Mean Squared Error）
是预测值与真实值偏差(标准差是与均值)的平方与观测次数n比值的平方根。衡量的是预测值与真实值之间的偏差，并且对数据中的异常值较为敏感

2.1均方根对数误差 RMSLE（Root Mean Squared Logarithmic Error）

1.RMSLE 惩罚欠预测大于过预测，适用于某些需要欠预测损失更大的场景，如预测共享单车需求。

假如真实值为 1000，若预测值为 600，那么 RMSE=400， RMSLE=0.510
假如真实值为 1000，若预测值为 1400， 那么 RMSE=400， RMSLE=0.336

可以看出来在 RMSE 相同的情况下，预测值比真实值小这种情况的 RMSLE 比较大，即对于预测值小这种情况惩罚较大。

2.如果预测的值的范围很大，RMSE 会被一些大的值主导。这样即使你很多小的值预测准了，但是有一个非常大的值预测的不准确，RMSE 就会很大。 相应的，如果另外一个比较差的算法对这一个大的值准确一些，但是很多小的值都有偏差，可能 RMSE 会比前一个小。先取 log 再求 RMSE，可以稍微解决这个问题。RMSE 一般对于固定的平均分布的预测值才合理。

3 R-Square（决定系数)!

（此处的R即相关系数，相关系数的平方就是决定系数R-Square。其中分母的y_mean是y_actual的mean。）

一般认为， R-Square越大越好。因为最佳情况下，分子（残差的平方）为0，R-Square等于1，它的作用是，把“误差平方和”这个好几万的数，变成 R-Square这个（一般来说）0~1的数，还能在只有这一套样本一个模型的情况下，知道预测结果大概准不准，大概有多准

4 中位数绝对误差 MAD

绝对中位差是一种统计离差的测量。而且，MAD是一种鲁棒统计量，比标准差更能适应数据集中的异常值。对于标准差，使用的是数据到均值的距离平方，所以大的偏差权重更大，异常值对结果也会产生重要影响。对于MAD，少量的异常值不会影响最终的结果。

由于MAD是一个比样本方差或者标准差更鲁棒的度量，它对于不存在均值或者方差的分布效果更好，比如柯西分布。
